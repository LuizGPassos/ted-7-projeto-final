# Imports e Dependências

```python
import pandas as pd
```

# Dataset

```python

df = pd.read_json('./dataset/movie_dataset.json')
df.head()
```

# Análise Exploratória (Para Limpeza de dados)

### Análise de colunas (Tipo e conteúdo)

```python
df.info()
```

### Validação de colunas numéricas

```python
df.describe()
```

### Exclui colunas com valores nulos

```python
df = df.drop(columns=['backdrop_path', 'belongs_to_collection', 'imdb_id', 'poster_path', 'homepage'])
df = df.dropna()
df.info()
```

### Pega primeiros valores de colunas tipo Array

```python
df['genre'] = df['genres'].apply(lambda x: x[0]['name'] if x else None)
df['company'] = df['production_companies'].apply(lambda x: x[0]['name'] if x else None)
df['country'] = df['production_countries'].apply(lambda x: x[0]['name'] if x else None)
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: x[0]['name'] if x else None)
df = df.drop(columns=['genres', 'production_companies', 'production_countries'])
df.head()
```

# Salva a tabela silver como CSV

```python
df.to_csv("./dataset/movie_dataset_silver.csv")
```
